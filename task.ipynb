```python
def friends(a):
    c = []
    for i in set(a):
        c.append(a.count(i))
        c.sort(reverse = True)
    for k in c:
        if k % min(c) == 0:
            n = k
            continue
        else:
            n = 1
            break
    return(n)
```

```python
cand  = ['a', 'a', 'a', 'b', 'b', 'b', 'b', 'c', 'c', 'c', 'c', 'c'] 
```

```python
friends(cand)
```
